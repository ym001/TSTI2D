```python
# -*- coding: utf-8 -*-
"""
Created on Fri Oct 18 23:09:34 2019

@author: marc kefer

Niveau : Terminale STI2D
Programme BO : Septembre 2020

Objectifs de ce programme :  
    - Comprendre la notion de demi-vie d'un isotope radioactif
    - Utiliser le modÃ¨le mathÃ©matique de la dÃ©croissance exponentielle
    - Comparer les demi-vies de deux isotopes radioactifs diffÃ©rents

CapacitÃ©s du BO travaillÃ©es :
    - Exploiter la dÃ©finition de la demi-vie d'une espÃ¨ce radioactive.
    - Comparer la dÃ©croissance radioactive de deux espÃ¨ces connaissant leurs
    demi-vies respectives.

ParamÃ¨tres modifiables :
         - N0 : Nombre initial de noyaux (commun aux 2 isotopes choisis)
         - Isotope1 et Isotope 2 : les noms des 2 isotopes
         - T_d1 et T_d2 : Les demi-vies respectives des 2 isotopes
"""
import numpy as np
import matplotlib.pyplot as plt

# Exemples de demi-vies: 
# Cobalt 60 :               5.2  ans
# Tritium :                 12.3 ans
# Strontium 90 :            28.1 ans
# CÃ©sium 137 :              30   ans
# AmÃ©ricium 241 :          432   ans
# Radium 226 :            1600   ans
# Carbone 14 :            5730   ans
# Plutonium 239 :        24110   ans
# Neptunium 237 :      2140000   ans
# Uranium 235 :        7038000   ans
# Iode 129 :          15700000   ans
# Uranium 238 :     4470000000   ans

## PARAMETRES MODIFIABLES ####################################################   
N0=6.4e23                     # Nombre initial de noyaux
# Noms des isotopes :
Isotope1="Tritium"
Isotope2="Uranium 235"
# Valeurs des demi-vies :
Td_1=12.3                     # En annÃ©es, pour l'isotope 1
Td_2=703800                   # En annÃ©es, pour l'isotope 2
###############################################################################
###############################################################################
def TracerGraphe(k,Isotope,Td,couleur):    
    print('##############################################')
    print("## Graphique",k,"#############################")
    print("## Nombre de noyaux - Isotope :", Isotope, "##")
    print('')
    print('Nombre initial de noyaux =', N0)
      
    plt.axis([0,5*Td , 0, 1.2*N0])
    plt.xticks([0,Td,2*Td,3*Td,4*Td,5*Td])
    plt.yticks([0,N0/8,N0/4,N0/2,N0])

    plt.text(1.1*Td, 0.9*N0, r'ModÃ¨le mathÃ©matique :',size='16')
    plt.text(1.5*Td, 0.8*N0, r'$N(t)=N_0 \times e^{-\lambda \times t}$',size='16', color=couleur)

    plt.grid([True])
    plt.ylabel('Nombre de noyaux restants',fontsize='15')
    plt.xlabel("AnnÃ©es",fontsize='15')
    plt.title(Isotope,fontsize='16')
    plt.show()
    print("Appuyer sur la touche 'EntrÃ©e' pour continuer.")
    input()

def TracerMultiGraphes(k,IsotopeA, IsotopeB,Td): 
    print('#####################################################')
    print("## Graphique ",k," ##################################")
    print("## Nombre de noyaux :", IsotopeA, "et", IsotopeB, "##")
    print('')
    print('Nombre initial de noyaux =', N0)
    plt.axis([0,5*Td , 0, 1.1*N0])
    plt.xticks([0,Td,2*Td,3*Td,4*Td,5*Td])
    plt.yticks([0,N0/8,N0/4,N0/2,N0])

    plt.text(Td, 0.8*N0, r'ModÃ¨les mathÃ©matiques :',size='15')
    TexteAffiche=Isotope1 +' : ' + r'$N_1(t)=N_0 \times e^{-\lambda_1 \times t}$'
    plt.text(Td, 0.7*N0, TexteAffiche,size='15',color='b')
    TexteAffiche=Isotope2 + ' : '+  r'$N_2(t)=N_0 \times e^{-\lambda_2 \times t}$'
    plt.text(Td, 0.6*N0, TexteAffiche,size='15',color='r')
    
    plt.legend()
    plt.grid([True])
    plt.ylabel('Nombre de noyaux restants',fontsize=12)
    plt.xlabel("DurÃ©e (ans)",fontsize=12)

    plt.show()
    
    print("Appuyer sur la touche 'EntrÃ©e' pour continuer.")
    input() 
   
# DÃ©finition des paramÃ¨tres graphiques ########################################
    # Constantes de temps (en annÃ©es) :
tau_1=Td_1/np.log(2)          # de l'isotope 1
tau_2=Td_2/np.log(2)          # de l'isotope 2
    # Liste des instants :
        # 1000 valeurs de t sur 5 demi-vies
t1=np.linspace(0,5*Td_1,1000)    # de l'isotope 1
t2=np.linspace(0,5*Td_2,1000)    # de l'isotope 2

        # 6 valeurs d'instants t, multiples entiers de la demi-vie 
t1_p=np.linspace(0,5*Td_1,6)    # de l'isotope 1
t2_p=np.linspace(0,5*Td_2,6)    # de l'isotope 2

# Liste des valeurs du nombre de noyaux restants :
    # Evolution du nombre de noyaux de l'isotope 1 :    
y11=N0*np.exp(-t1/tau_1)         # sur 5 demi-vies de l'isotope 1 
y11_p=N0*np.exp(-t1_p/tau_1)

y12=N0*np.exp(-t2/tau_1)         # sur 5 demi-vies de l'isotope 2
y12_p=N0*np.exp(-t2_p/tau_1)    

   # Evolution du nombre de noyaux de l'isotope 2 :
y21=N0*np.exp(-t1/tau_2)        # sur 5 demi-vies de l'isotope 1 
y21_p=N0*np.exp(-t1_p/tau_2)

y22=N0*np.exp(-t2/tau_2)        # sur 5 demi-vies de l'isotope 2
y22_p=N0*np.exp(-t2_p/tau_2)

## ReprÃ©sentations graphiques #################################################
## Evolution du nombre de noyaux de l'isotope 1
i=1
plt.plot(t1,y11,'-b')
plt.plot(t1_p,y11_p,'ob')
TracerGraphe(i,Isotope1,Td_1,'b')

## Evolution du nombre de noyaux de l'isotope 2
i=2
plt.plot(t2,y22,'-r')
plt.plot(t2_p,y22_p ,'or')
TracerGraphe(i,Isotope2,Td_2,'r')

## Evolution des nombres de noyaux des deux isotopes (sur 5 demi-vie de l'isotope 2)
i=3
plt.plot(t2,y12 ,'-b', label=Isotope1,linewidth=3)
plt.plot(t2_p,y22_p,'or')
plt.plot(t2,y22,'-r', label=Isotope2)
TracerMultiGraphes(i,Isotope1, Isotope2,Td_2)

## Evolution des nombres de noyaux des deux isotopes (sur 5 demi-vie de l'isotope 1)
i=4
plt.plot(t1,y11 ,'-b', label=Isotope1)
plt.plot(t1_p,y11_p,'ob')
plt.plot(t1,y21,'-r', label=Isotope2)
TracerMultiGraphes(i,Isotope1, Isotope2,Td_1)
###############################################################################
print('Fin du programme')
```
